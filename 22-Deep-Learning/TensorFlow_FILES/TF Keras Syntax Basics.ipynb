```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("DATA/fake_reg.csv")
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1', 'feature2']].values
y = df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
help(MinMaxScaler)
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

# Creating and Training the Model

```python
# Import
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense
```

```python
help(Sequential)
```

```python
# Method 1 of creating a NN
model = Sequential([Dense(4, activation = 'relu'),
                   Dense(2, activation = 'relu'),
                   Dense(1)])
```

```python
# Method 2: This method is preferred
model = Sequential()

model.add(Dense(4, activation = 'relu'))
model.add(Dense(2, activation = 'relu'))
model.add(Dense(1))
```

```python
# Actual notes
model = Sequential()

model.add(Dense(4, activation = 'relu'))
model.add(Dense(4, activation = 'relu'))
model.add(Dense(4, activation = 'relu'))

model.add(Dense(1))

model.compile(optimizer = 'rmsprop', loss = 'mse')
```

```python
model.fit(x = X_train, y = y_train, epochs = 250)
```

```python
model.history.history
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

## Model Evaluation

```python
# This simply evaluates the value that the loss "converges" to in terms of mse
model.evaluate(X_test, y_test, verbose = 0)
```

```python
model.evaluate(X_train, y_train, verbose = 0)
```

```python
# Now to actually test the predictions
test_predictions = model.predict(X_test)
```

```python
test_predictions = pd.Series(test_predictions.reshape(300,))
pred_df = pd.DataFrame(y_test, columns = ['Test True Y'])
pred_df = pd.concat([pred_df, test_predictions], axis = 1)
```

```python
pred_df.columns = ['Test True Y', 'Model Predictions']
pred_df
```

```python
sns.scatterplot(data = pred_df, x = 'Test True Y', y = "Model Predictions")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(pred_df["Test True Y"], pred_df["Model Predictions"])
```

```python
mean_squared_error(pred_df["Test True Y"], pred_df["Model Predictions"])
```
